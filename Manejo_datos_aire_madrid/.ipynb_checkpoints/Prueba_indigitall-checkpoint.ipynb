# 1. Análisis de los conjuntos de datos obtenidos del portal de madrid en formato CSV
Se ha descargado un ZIP con los distintos conjuntos de datos por hora y en tiempo real asociados al año 2018 sobre la calidad del aire en Madrid. Cada uno de los conjuntos de datos representa la calidad del aire para un mes del 2018, para cada hora de cada día y se presenta en 3 tipos de formatos: txt, csv y xml. He decidido utilizar los que se encuentran en formato CSV pues conozco librerías de Python que facilitan su carga como dataframes de Pandas, lo cual facilita también su manipulación posterior (ETL).

**A continuación se expone una explicación del contenido del conjunto de datos (CSV), describiendo sus cabeceras, para comprender bien la información contenida. Para ello he utilizado la información expuesta por el portal de Madrid:
https://datos.madrid.es/FWProjects/egob/Catalogo/MedioAmbiente/Aire/Ficheros/Interprete_ficheros_%20calidad_%20del_%20aire_global.pdf**

- **Provincia**: Código de provincia donde se encuentra la estación.
- **Municipio**: Código de municipio donde se ubica la estación.
- **Estacion**: Código que identifica la estación.
- **Magnitud**: Campo que identifica la magnitud medida, por ejemplo, el código 08 u 8 se corresponde con el Dióxido de Nitrógeno (NO2), la magnitud que nos interesa.
- **Punto_muestreo**: Se conforma con el código completo de la estación (provincia, municipio y estacion) junto con la magnitud y la técnica de muestreo separadas por "_". Por ejemplo: Si provincia tiene código 28, municipio 79, estacion 4, magnitud 1 y la técnica de muestreo es la 38: 28079004_1_38. En formato TXT la técnica también se incluye como cabecera, aquí se deduce de este campo. Cabe destacar que el municipio y la estación se identifican con un cópdigo compuesto por 3 dígitos, de ahí que se emplee 079 y 004 respectivamente. Para el ejemplo dado, el identificador 28079004 se asocia con la estación de Plaza España.
- **Ano**: Año de la medición. Para este informe, solo se contempla 2018.
- **Mes**: Mes de la medición, indicado en formato entero. Por ejemplo, 04 corresponde con Abril.
- **Dia**: Día de la medición, en formato entero. 
- **HMN** (M puede ser 0, 1, o 2, N es un entero comprendido entre 1 y 4): Representa el dato tomado a la hora dada por M y N. Por ejemplo H01 se corresponde con la 1 de la mañana o H24 con las 00:00. Lo coloco en este formato para indicar que hay múltiples columnas desde H01 a H24.
- **VMN** (M puede ser 0, 1 o 2,N es un entero comprendido entre 1 y 4): Es el código de validación del dato recogido en la hora indicada por el campo anterior. El formato es el mismo cambiando solo la H por la V. Lo coloco en este formato para indicar que hay múltiples columnas desde V01 a V24. Solo son válidos los datos con el código de validación "V".



# 2. Objetivo 1. Tratamiento de los datos y predicciones
## - En primer lugar se van a manipular los datos. 
Se van a unir los archivos CSV correspondientes a cada mes en un único fichero. Se van a extraer solo los datos asociados a NO2 (magnitud 8). Se podrían aplicar otras transformaciones como filtrar únicamente los datos validados, pero esta información también puede reultar útil para monitorizar el estado de las estaciones o las técnicas aplicadas y sacar conclusiones. 

```python
import numpy as np
import pandas as pd

#los ficheros tienen formato <MES>_mo18.csv, donde <MES> toma un valor de los de la siguiente tupla (inmutable)
files_header_names = ("ene", "feb", "mar","abr","may","jun", "jul", "ago", "sep", "oct", "nov", "dic")
# Creo una lista con los dataframes para unirlos a continuación en uno solo
datasets = []
for h in files_header_names:
    file = "Datasets/"+h+"_mo18.csv"
    ds = pd.read_csv(file, sep=";")
    datasets.append(ds)
print(len(datasets))
# Uno los dataframes, pues tienen cabecera comunes, tomando el primer dataframe de la lista como punto de partida
df = datasets[0]
for d in datasets[1:]:
    df = pd.concat([df,d], axis = 0)

# Compruebo que se han cargado todos los meses
print(df[["MES"]].drop_duplicates(inplace = False, keep = "first"))

#Compruebo información sobre el dataframe resultante
print(df.info())
print(df.shape)
```

```python
# Filtro únicamente por magnitud = 8, que se corresponde con NO2
print(df.columns)
dfNO2 = df.loc[df["MAGNITUD"] == 8]
print(dfNO2.head())
print(len(dfNO2))

```

**A continuación se van a cargar y transformar los datasets madrid_hourly_temperatures_2018.csv y madrid_air_quality_stations.csv como dataframes de Pandas. El primero contiene información diaria y horaria de la temeperatura en Madrid,mientras que el segundo contiene información sobre las estaciones, como la altitud o coordenadas. Ambos son útiles para explorar los datos en el punto siguiente.**

```python
madrid_hourly_temperatures_2018 = pd.read_csv("Datasets/madrid_hourly_temperatures_2018.csv", sep = ",")
madrid_air_quality_stations = pd.read_csv("Datasets/madrid_air_quality_stations.csv", sep = ",")
# Transformo madrid_hourly_temperatures_2018 para obtener una temperatura media por día a partir de las temperaturas horarias. 
# Se usa reset_index para restablecer el índice por el nuevo dataframe resultante.
madrid_hourly_temperatures_2018_mean_per_hour = madrid_hourly_temperatures_2018.groupby('date')['temp'].mean().round(2).reset_index()
madrid_hourly_temperatures_2018_mean_per_hour.head()
```

```python
# Dataframe con valor medio de los datos por estacion
print(dfNO2.columns[9::2])
media_datos_estacion = dfNO2.groupby("ESTACION")[dfNO2.columns[8::2]].mean().reset_index()
media_datos_estacion['MEDIA'] = media_datos_estacion.mean(axis=1).round(4)
print(media_datos_estacion.columns)
media_datos_estacion = media_datos_estacion[media_datos_estacion.columns[[0,25]]]
media_datos_estacion.head()
```

```python
# Dataframe con numero de datos no validados por estación

media_datos_no_validados_estacion = dfNO2.copy()
media_datos_no_validados_estacion['NUM_ERRORES'] = media_datos_no_validados_estacion.apply(lambda x: x[9::2].tolist().count('N'), axis=1)
media_datos_no_validados_estacion = media_datos_no_validados_estacion[media_datos_no_validados_estacion["NUM_ERRORES"] != 0]
#print(media_datos_no_validados_estacion[media_datos_no_validados_estacion["ESTACION"] == 4])
#print(media_datos_no_validados_estacion.iloc[0])
media_datos_no_validados_estacion = media_datos_no_validados_estacion.groupby("ESTACION")["NUM_ERRORES"].sum().reset_index()
media_datos_no_validados_estacion.head()
#media_datos_no_validados_estacion.to_csv('media_datos_no_validados_estacion.csv',index = False, sep = ";")
```

```python
# Por meses valor máximo medido
max_data_value_mes = dfNO2.copy()
max_data_value_mes = max_data_value_mes.groupby('MES').max()[max_data_value_mes.columns[8::2]].max(axis=1).reset_index()
max_data_value_mes.columns = ["MES","MAXIMO"]
max_data_value_mes.head()
#max_data_value_mes.to_csv('max_data_value_mes.csv',index = False, sep = ";")
```

```python
# Media datos medidos en función de la altitud de la estación
datos_segun_altitud_station = dfNO2.copy()
datos_segun_altitud_station["station"] = list(map(lambda x: int(x.split("_")[0]), dfNO2["PUNTO_MUESTREO"]))
datos_segun_altitud_station = pd.merge(datos_segun_altitud_station, madrid_air_quality_stations, on='station', how='inner')

datos_segun_altitud_station = datos_segun_altitud_station.groupby("altitude")[datos_segun_altitud_station.columns[8:56:2]].mean().reset_index()
datos_segun_altitud_station['MEDIA'] = datos_segun_altitud_station.mean(axis=1)
datos_segun_altitud_station = datos_segun_altitud_station[datos_segun_altitud_station.columns[[0,25]]]
datos_segun_altitud_station.head()
#datos_segun_altitud_station.to_csv('datos_segun_altitud_station.csv',index = False, sep = ";")
```

```python
# Media diaria de temperatura en base a las media diaria de mediciones de datos NO2. Uno madrid_hourly_temperatures_2018_mean_per_hour con dfNO2 por date.

media_datos_por_temp = dfNO2.copy()
media_datos_por_temp['date'] = media_datos_por_temp[['ANO', 'MES', 'DIA']].apply(lambda x: '-'.join(map(lambda y: str(y).zfill(2), x)), axis=1)
media_datos_por_temp = pd.merge(media_datos_por_temp, madrid_hourly_temperatures_2018_mean_per_hour, on='date', how='inner')

media_datos_por_temp = media_datos_por_temp.iloc[:,8:]

#madrid_hourly_temperatures_2018_mean_per_hour
media_datos_por_temp = media_datos_por_temp.groupby("temp")[media_datos_por_temp.columns[0:48:2]].mean().reset_index()

media_datos_por_temp['MEDIA'] = media_datos_por_temp[media_datos_por_temp.columns[1:]].mean(axis=1)
media_datos_por_temp = media_datos_por_temp[media_datos_por_temp.columns[[0,25]]]

media_datos_por_temp.head()
#media_datos_por_temp.to_csv('media_datos_por_temp.csv',index = False, sep = ";")
```

```python
#Media datos medidos por fecha
media_datos_por_fecha = dfNO2.copy()
media_datos_por_fecha['date'] = media_datos_por_fecha[['ANO', 'MES', 'DIA']].apply(lambda x: '-'.join(map(lambda y: str(y).zfill(2), x)), axis=1)
media_datos_por_fecha = pd.merge(media_datos_por_fecha, madrid_hourly_temperatures_2018_mean_per_hour, on='date', how='inner')
media_datos_por_fecha.columns[8:56:2]
media_datos_por_fecha = media_datos_por_fecha.groupby("date")[media_datos_por_fecha.columns[8:56:2]].mean().reset_index()
print(media_datos_por_fecha.columns)
media_datos_por_fecha['MEDIA'] = media_datos_por_fecha[media_datos_por_fecha.columns[1:]].mean(axis=1)
media_datos_por_fecha = media_datos_por_fecha[media_datos_por_fecha.columns[[0,25]]]
media_datos_por_fecha.head()
```

## - Modelo de predicción a 3 días vista de los niveles de NO2

A la hora de efectuar predicciones, en este caso de los niveles de NO2 a 3 días vista conviene indentificar variables que influyan en los datos que se van a estudiar para realizar la predicción e identificar posibles patrones en los datos que sirven de muestra. Para ello se pueden agregar nuevas variables. Como se ha visto, anteriormente se ha integrado el dataframe inicial con otros que guardan datos de temperaturas o de la altitud de la estación. Se ha observado que existe una relación entre los datos medidos y la temperatura, de forma que al incrementar esta, el valor medido es menor. Con ello se pueden incluir valores relacionados con la meteorología, como temperatura, humedad, presión atmosférica y velocidad o dirección del viento. Estos factores suelen depender de la época del año en la que nos encontramos, y además, es algo fundamental para la predicción tener en cuenta la fecha, pues se pretende predecir a 3 días vista, por lo que mes, día y año son variables importantes a tener en cuenta. En función de cuanto se quiera precisar la predicción, se podría incluir también la variable hora. Por último, no es lo mismo realizar una predicción en madrid centro que en otra ubicación, con lo que el punto de muestreo también formará parte del conjunto de muestras.  

En cuanto a la profundidad de datos históricos, en el portal de Madrid, disponen de datos desde el 2001. Es importante contar con un gran abanico de datos para poder efectuar el entrenamiento, no obstante, en base a las variables elegidas, hay que tener en cuenta que en los últimos años el índice de contaminación en el aire a aumentado así como la temperatura. Esto es algo que podría marcar el entrenamiento del modelo, por lo que se podrían emplear datos, por ejemplo de los últimos 10 años, periodo en el que quizá la diferencia no es tan grande que si hablamos de más de 20 años. 

El primer paso será preparar los datos y limpiarlos, he mencionado datos históricos de los últimos 10 años, pero habría que verificar que tienen los campos que interesan para el entrenamiento. Tras sus estandarización habrá que generar conjuntos de datos de entrenamiento, validación y prueba.

Existen distintos modelos predictivos, como es el caso de modelos de clasificación, clustering, regresión, de series de tiempo, árboles de decisión o redes neuronales. Los modelos de clasificación permiten categorizar información en base a datos históricos, los de clustering permiten agrupar objetos en base a características (por ejemplo segmentar clientes). Los modelos de series temporales se emplean para analizar datos a lo largo del tiempo con el fin de identificar patrones y tendencias. A su vez, las redes neuronales, permiten por ejemplo reconocer patrones en gtandes conjuntos de datos. Los modelos de regresión permiten identificar como unas variables independientes influyen en una dependiente. En el caso de los árboles de decisión, se pueden emplear tanto para clasificiación como para regresión. Teniendo en cuenta esto, para el problema habría que decidir entre utilizar un modelo de regresión o de series temporales. Se quiere ver como una serie de variables independientes (temperatura, hora del día...) influyen sobre una dependiente (nivel de NO2) para poder efectuar una predicción futura. Esto se ajusta bien a un modelo de regresión, sin embargo, este tipo de modelos no procesa bien la complejidad temporal de los datos, es decir, factores como la estacionalidad, ciclos o tendencias temporales, por ejemplo, de los niveles de NO2 no se tratarían correctamente. Con ello el modelo elegido es un **modelo de series temporales** .



# 3. Objetivo 2. Arquitectura planteada para el almacenamiento, limpieza, explotación y presentación de los datos.

**La arquitectura a diseñar debe cumplir varios requisitos:**
1. Carga: 
    - Cargas de info masivas (puntuales): carga inicial y esporádica para corregir información en tiempo real. Esto puede implicar la existencia de un data lake que almacene datos en bruto (también menciona que se usan para corregir información capturada en tiempo real) o un data warehouse, que podría alimentarse del anterior, con datos estructurados.
    - Cargas incrementales a partir de datos en tiempo real (cada hora) de contaminación y datos de temperaturas obtenidos por APIs de datos meteorológicos. Aquí se podría obtener información estructurada y semiestructurada (si se almacenan los ficheros con la información en formatos como CSV por ejemplo). Esta información podría procesarse en tiempo real para sacar conclusiones o disparar eventos, como una alerta por contaminación o para monitorizar algún dispositivo.
2.  Explotación:
    - API
    - Portal Web
    - Modelo predictivo
    - Ad hoc
3. Volumen: Desde año 2001
4. Usuarios: 500 - 1000 usuarios. No es un número excesivamente grande de usuarios, pero convendría por ejemplo utilizar servicios en nube, incluso serveless que puedan manejar escalabilidad frente a un gran número de peticiones usando por ejemplo balanceadores y contenedores.
   
**A continuación se procede con el diseño de la arquitectura**

La arquitectura que se plantea esta inspirada en la arquitectura lambda, pues se pretende combinar un procesamiento batch con otro streaming. Esto es debido a que, pese a que por lo general las cargas se realizan en tiempo real, no siempre se procesan al instante, si no que se almacena la información, por lo general procesada, eso es cierto, para posteriormente ser consultada (por ejemplo, explotación ad hoc). La funcionalidad streaming puede resultar útil para funcionalidades como la monitorización en tiempo real, que dispare algún tipo de evento.

He planteado 2 tipos de almacenamiento principales: data lake y data warehouse. Podría llegar a ser excesivo pues no se espera una gran diversidad en cuanto a los tipos de fuentes de información desde los que se cargan los datos, quienes por lo general son estructurados y semiestructurados. La idea del data lake radica en almacenar la maxima información posible, y puede resultar útil para las cargas masivas de datos así como de disponer datos histróricos permitiendo contar con una analítica del tipo diagnóstica y predictiva y no solo descriptiva. Con ello el sistema contará con un datalake que nutrirá a un alamacén de datos que alimenta a su vez distintos datamarts orientados a negocio, donde, por ejemplo, los científicos de datos pueden realiza predicciones. Aquí se reflejan procesos ETL y ELT.

Para la implementación de la arquitectura se ha planteado utilizar Hadoop, pues ofrece un entorno con muchas herramientas orientadas al Big Data, tanto para las fases de extracción, almacenamiento y explotación. Además permite implementar el data lake permitiendo el almacenamiento y procesamiento distribuido de los datos, gracias a sus sistema de ficheros distribuido HDFS y el framework para el procesamiento distribuido basado en Map-Reduce. 

A continuación se presentan las **capas** que tendrá la arquietctura y las **tecnologías** o herramientas que se pueden adoptar para su implementación:

- **Capa funcional**
En esta capa se incluyen las distintas capas destinadas a la ingesta, almacenamiento, procesamiento, explotación, visualización y coordinación del flujo de trabajo.

    - **Capa de ingesta.**
Se encarga de la carga en el ecosistema de Hadoop de datos procedentes de las distintas fuentes mencionadas. Para ello se utilizan herramientas como Sqoop o Flume. Sqoop se encarga de transferir datos entre entornos de datos relacionales al HDFS. Flume por su parte admite distintas fuentes de datos, incluyendo las no estructuradas. Además, se enfoca en la recopilación de datos en tiempo real. Para la carga de los datos de una API meteorológica se pueden usar scripts en Python que efectúen las consultas necesarias, periódicamente si se desea.

- **Capa de almacenamiento.**
En esta capa se almacenan los correspondientes datos en bruto o procesados. Está compuesta por un lago de datos, implementado mediante el HDFS y un almacén de datos. El lago de datos se compone por 3 capas, la primera almacena datos en bruto tal cual llegan de la ingesta, la segunda datos estandarizados, que son datos que han pasado por procesos de filtrado o limpieza para  facilitar su uso. La tercera son datos procesados, consiste en información consolidada lista para ser  utilizada. El lago a su vez nutre al almacén de datos, que guarda cierta información estructurada y a los datamarts, repositorios que centralizan la información a las distintas áreas de negocio. Para implementar el almacén y los datamarts existen alternativas en la nube como son Amazon Redshift, Google BigQuery. Big Query tiene ventaja en cuanto a la capacidad de acceder a volúmenes grandes de datos y realizar  consultas rápidas, pues separa almacenamiento de las consultas. Se ha decidio utilizar este último, quien ofrece servicios para ETL a través de Google Cloud Dataflow y se puede integrar con Google Cloud Platform.

    - **Capa de procesamiento.**
Esta capa se encarga de realizar distintas operaciones sobre los diferentes datos almacenados o flujos de datos. Aquí se diferencian dos tipos de procesamiento, batch y streaming. Para el primer caso existen herramientas como Apache Pig y Apache Spark. Hay que tener en cuenta que la segunda además de disponer de un motor de ejecución más optimizado admite varios lenguajes de programación (Python, R, Java), ofrece una API de alto nivel y proporciona un mejor rendimiento que Pig, quien utiliza un lenguaje propio denominado Pig Latin. Además, se integra perfectamente con herramientas como Hive o Hbase. Por otro lado, para el procesamiento streaming existen herramientas como Spark Streaming y Apache Storm, siendo más complejo de manejar este último. Se ha decidido emplear tecnologías Spark en ambos casos.

    - **Capa de explotación y acceso.**
A través de esta capa se interactúa con los distintos datos almacenados a través de múltiples  herramientas. Para el procesamiento de datos SQL se contemplan Hive e Impala. El primero es más adecuado para procesos que no requieren respuestas en tiempo real, mientras que el segundo se emplea en análisis en tiempo real, donde la consistencia no es tan crítica. En el caso de tener que acceder a datos No SQL se podrían usar HBase o MongoDB. Además, tal y como se ha mencionado, aquí se incluirían una o más API para ofrecer la información a los usuarios finales.
 
    - **Capa de visualización.**
Los usuarios finales, los analistas de datos, quienes utilizan herramientas para generar gráficos y cuadros de mandos, o científicos de datos acceden a la información a través de esta capa. Las herramientas planteadas son Power Bi y Tableau. Esta última es recomendada para grandes empresas, ya que Power Bi no admite archivos de más de 1GB. Aquí se incluye también el portal web que, a través de gráficas y cuadros de mando, muestra información a los usuarios autorizados.

    - **Capa de coordinación.**
Capa que planifica cuando lanzar los distintos trabajos de procesamiento de datos, facilitando la orquestación y ejecución de las tareas en el clúster de Hadoop. Aquí destaca el uso de Oozie, un sistema de orquestación de flujos de trabajo para la gestión de tareas en entornos de Hadoop. Por ejemplo,  coordina trabajos MapReduce, de Pig o de Hive.

- **Capa de gestión de recursos del cluster**
Esta capa asigna y administra los recursos disponibles en el clúster. Aquí se incluye Apache Yarn, que proporciona un planificador agnóstico a los trabajos que se encuentran en ejecución en el clúster.
 
- **Capa Hadoop Common**
Esta capa contiene el conjunto de bibliotecas comunes empleadas por otros módulos de Hadoop.
  
- **Capa del hardware**
En esta capa se encuentra el hardware correspondiente entre el que estaría el clúster de nodos de 
Hadoop.

Tras esto, se plantea una **explicación del funcionamiento** de la arquitectura en alto nivel. 

Los datos de las distintas fuentes entran en la capa de ingesta. Por un lado los estructurados son cargados 
por Sqoop, mientras que del resto, si los hubiese, se encarga Flume. Sqoop ingresa los datos en el ecosistema Hadoop, 
dirigiéndolos al lago de datos implementado por HDFS, donde se cargan en bruto para posteriormente ser procesados convenientemente, mediante herramientas como Spark o Mahout, previamente explicadas. Tras ser procesados, pueden almacenarse en una nueva capa del lago. Es lo que se conoce como ELT. Por otro lado, Flume hace lo mismo con las fuentes de datos restantes, cargando en el lago de datos únicamente los datos que se considere necesario mantener, pues el resto se procesarán en tiempo real a través de Spark Streaming. Este procesado en tiempo real permite 
también almacenar los datos si fuese necesario. Los datos almacenados en el lago pueden nutrir el almacén de datos y este a su vez los datamarts, los cuales se pueden conectar a herramientas de inteligencia del negocio y ser explotados por analistas de datos. La capa de coordinación se va a encargar, con Oozie, de gestionar todo el flujo de trabajo de procesamiento de datos.

El acceso y la explotación se puede efectuar de distintas formas, ya sea mediante APIs o herramientas como Hive o Impala. Por último una capa de visualización permite mostrar y analizar los datos, crear métricas, cuadros de mandos, predicciones o ser utilizados en aplicaciones. Se considera también un área de sandboxes, donde los científicos de datos acceden al lago de datos en entornos aislados para realizar localizar patrones y realizar predicciones.

Por debajo se pueden ver 3 capas correspondientes al hardware, esto es, el clúster y otros elementos necesarios, las bibliotecas del entorno de Hadoop y el framework de procesamiento distribuido basado en Map Reduce, gestionado por Yarn.
